```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("Metadata_Country_API_SP.POP.TOTL_DS2_en_csv_v2_84031.csv")
```

```python
df.head(10)
```

```python
income_df = df[['Country Code','IncomeGroup']]
```

```python
income_df = income_df.dropna(subset=['IncomeGroup'])
```

```python
income_counts = income_df['IncomeGroup'].value_counts()
```

```python
plt.figure(figsize=(10,6))
income_counts.plot(kind='bar' , color='brown')
plt.xlabel('Income Group')
plt.ylabel('Number of countries')
plt.title('Distribution of Countries by Income Group')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
